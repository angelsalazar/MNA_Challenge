### Normalization

### Binary Values

the following columns:

* CUMPLE_CON_ALC          
* CUMPLE_CON_COND         
* CUMPLE_CON_SDT_ra       
* CUMPLE_CON_SDT_salin    
* CUMPLE_CON_FLUO         
* CUMPLE_CON_DUR          
* CUMPLE_CON_CF           
* CUMPLE_CON_NO3          
* CUMPLE_CON_AS           
* CUMPLE_CON_CD           
* CUMPLE_CON_CR           
* CUMPLE_CON_HG           
* CUMPLE_CON_PB           
* CUMPLE_CON_MN           
* CUMPLE_CON_FE  

have a binary value ("SI"/"NO"). These will be casted as numbers where "SI" is 1 and "NO", 0.

```python
cols = [
  'CUMPLE_CON_ALC',          
  'CUMPLE_CON_COND',        
  'CUMPLE_CON_SDT_ra',       
  'CUMPLE_CON_SDT_salin',    
  'CUMPLE_CON_FLUO',        
  'CUMPLE_CON_DUR',          
  'CUMPLE_CON_CF',           
  'CUMPLE_CON_NO3',          
  'CUMPLE_CON_AS',           
  'CUMPLE_CON_CD',           
  'CUMPLE_CON_CR',           
  'CUMPLE_CON_HG',           
  'CUMPLE_CON_PB',           
  'CUMPLE_CON_MN',           
  'CUMPLE_CON_FE'  
]
for col in cols:
  cleanedDf[col].replace({"SI": 1, "NO": 0}, inplace=True) 

cleanedDf.head()
```

In a closer look at the dataset, the following columns:

* SDT_M_mg/L
* FLUORUROS_mg/L
* DUR_mg/L
* COLI_FEC_NMP/100_mL
* N_NO3_mg/L
* AS_TOT_mg/L
* CD_TOT_mg/L
* CR_TOT_mg/L
* HG_TOT_mg/L
* PB_TOT_mg/L
* MN_TOT_mg/L
* FE_TOT_mg/L

were identified to have mixed data types as values (strings and floats). The string values indicate that level of a chemical did not exceed a certain threshold and the real ones were omitted.

In order to normalize the string values, they will be replaced with a random number that does not overpass the indicated threshold.

```python
# utilitary function
def createThresholdParser(column):
  def thresholdParser(row):
    value = row[column]
    if isinstance(value, str) and '<' in value:
      threshold = value.strip().replace('<', '')
      threshold = float(threshold)
      value = np.random.uniform(0, threshold, 1)[0]
    return value
  return thresholdParser
```

```python
# defining columns that will get the 
# string values replace with a random number
waterChemicalCols = [
    'SDT_M_mg/L',
    'FLUORUROS_mg/L',
    'DUR_mg/L',
    'COLI_FEC_NMP/100_mL',
    'N_NO3_mg/L',
    'AS_TOT_mg/L',
    'CD_TOT_mg/L',
    'CR_TOT_mg/L',
    'HG_TOT_mg/L',
    'PB_TOT_mg/L',
    'MN_TOT_mg/L',
    'FE_TOT_mg/L'
]

for column in waterChemicalCols:
  # applying custom thresholdParser
  # and saving the replaced values in the target column
  cleanedDf[column] = cleanedDf.apply(createThresholdParser(column), axis = 1)
  # casting updated column as float
  cleanedDf[column] = cleanedDf[column].astype(np.float64)
```

## Clustering


```python
# importing Kmeans library
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
```

```python
# defining independant features
clusteringDf = cleanedDf[["LONGITUD", "LATITUD"]]
clusteringDf.head()
```

```python
# plotting coords
sns.scatterplot(data=clusteringDf, x="LONGITUD", y="LATITUD")
```

### Computing optimal number of clusters

```python
# the optimal number of clusters 
# gets computed through the elbow method

# Instantiate the clustering model and visualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=(2,10))
 # Fit the data to the visualizer
visualizer.fit(clusteringDf)       
# ploting KElbow
visualizer.show()                 
```

The illustration displayed above shows that the optimal number of clusters is 4.

### Model training


```python
# the value of 4 gets assgined based on the result
# provided by the elbow visualizer
OPTIMAL_N_CLUSTERS = 4 

# creating instance of KMeans model
cluster = KMeans(n_clusters = OPTIMAL_N_CLUSTERS)

# training model
cluster.fit(clusteringDf)

# predicting model
y_pred = cluster.predict(clusteringDf)
```

```python
clusteringDf['K_MEANS_PRED'] = y_pred
```

```python
clusteringDf['SEMAFORO'] = cleanedDf['SEMAFORO']
```

```python
clusteringDf.head()
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,7)) 

sns.scatterplot(
    data=clusteringDf, 
    x="LONGITUD", 
    y="LATITUD", 
    hue="K_MEANS_PRED",
    ax = ax1
).set(title='KMeans clustering')
sns.scatterplot(
    data = clusteringDf,
    x = "LONGITUD",
    y = "LATITUD",
    hue = "SEMAFORO",
    ax = ax2,
    palette=dict(Verde='#34D399', Amarillo='#FCD34D', Rojo='#F43F5E')
).set(title='Semaphore label')
```

Based on the illustrations shown above, it can be concluded that there is no relationship between the entries' coordinates and water quality. KMeans clustering result does not match the semaphore-coloring result. However, cluster number 3 has a certain approximation to the semaphore-coloring result located in the bottom-right corner.
