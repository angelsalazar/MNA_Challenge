## Data Cleansing

### Identifying Missing data

```python
# cloning datafram
cleanedDf = df.copy()
```

```python
# display a total count of rows having missing data per column
cleanedDf.isna().sum()
```

Based on the summary displayed above, the column "SDT_mg/L" has no data, therefore it is safe to drop it

```python
# dropping column "SDT_mg/L"
cleanedDf.drop('SDT_mg/L', axis = 1, inplace = True)
```

```python
# Column "PERIODO" gets dropped
# since all entries were recorded in 2020
cleanedDf.drop('PERIODO', axis = 1, inplace = True)
```

The following columns:

* ALC_mg/L
* CALIDAD_ALC
* CONDUCT_mS/cm
* SDT_M_mg/L  
* CALIDAD_CONDUC
* SDT_M_mg/L
* CALIDAD_SDT_ra 
* CALIDAD_SDT_salin
* DUR_mg/L   
* CALIDAD_DUR      
* N_NO3_mg/L         
* CALIDAD_N_NO3

the columns have at most 6 rows without data. Since the total of rows having missing information is not significant, the fastest and most straightforward approach is to remove them from the dataframe.

```python
# defining columns to clean up
targetColumns = [
  'ALC_mg/L',
  'CALIDAD_ALC',
  'CONDUCT_mS/cm',
  'SDT_M_mg/L',
  'CALIDAD_CONDUC',
  'SDT_M_mg/L',
  'CALIDAD_SDT_ra',
  'CALIDAD_SDT_salin',
  'DUR_mg/L',
  'CALIDAD_DUR',
  'N_NO3_mg/L',
  'CALIDAD_N_NO3'
]

cleanedDf.dropna(subset = targetColumns, inplace = True)
```

For the categorical column "CONTAMINANTES", it can be assumed that some rows have no values because the water quality is the best. Instead keeping the missing values, they will be replaced by N/A (Not Applied)

```python
# replacing missing values
cleanedDf['CONTAMINANTES'].fillna('NO APLICA', inplace=True)
```

```python
# displaying total of missing values per column
cleanedDf.isna().sum()
```

### Box plots and mean

The water quality is denoted based on the following columns:

* CUMPLE_CON_ALC          
* CUMPLE_CON_COND         
* CUMPLE_CON_SDT_ra       
* CUMPLE_CON_SDT_salin    
* CUMPLE_CON_FLUO         
* CUMPLE_CON_DUR          
* CUMPLE_CON_CF           
* CUMPLE_CON_NO3          
* CUMPLE_CON_AS           
* CUMPLE_CON_CD           
* CUMPLE_CON_CR           
* CUMPLE_CON_HG           
* CUMPLE_CON_PB           
* CUMPLE_CON_MN           
* CUMPLE_CON_FE  

if an entry has 1s in all of the fields, that means that the water quality is the best. In order to identify outliers, each entry will get a score (sum of all values of the fields listed above). This value will be used to compute the mean and plot a boxplot illustration to easily visualize outliers.

```python
# creating a copy
visualizationDf = cleanedDf.copy()
```

```python
# utilitary function
def createAdder(columnNames):
  def customAdder(row):
    total = 0;
    for col in columnNames:
      total += row[col]
    return total
  return customAdder;
```

```python
# creating custom adder
computeWaterScore = createAdder([
  'CUMPLE_CON_ALC',          
  'CUMPLE_CON_COND',        
  'CUMPLE_CON_SDT_ra',       
  'CUMPLE_CON_SDT_salin',    
  'CUMPLE_CON_FLUO',        
  'CUMPLE_CON_DUR',          
  'CUMPLE_CON_CF',           
  'CUMPLE_CON_NO3',          
  'CUMPLE_CON_AS',           
  'CUMPLE_CON_CD',           
  'CUMPLE_CON_CR',           
  'CUMPLE_CON_HG',           
  'CUMPLE_CON_PB',           
  'CUMPLE_CON_MN',           
  'CUMPLE_CON_FE'
])
```

```python
# computing score for each entry
visualizationDf['Score'] = visualizationDf.apply(computeWaterScore, axis=1)
```

```python
visualizationDf.head()
```

```python
# ploting boxplot and mean
fig, ax = plt.subplots(figsize=(15,15)) 
sns.boxplot(data = visualizationDf[
    ['Score']
  ],
  notch=True, 
  showcaps=False,
  flierprops={"marker": "x"},
  medianprops={"color": "coral"},
  ax = ax
)


```

## Clustering


### Model training


```python
# importing Kmeans library
from sklearn.cluster import KMeans
```

```python
# defining independant features
clusteringDf = cleanedDf[["LONGITUD", "LATITUD"]]
clusteringDf.head()
```

```python
# plotting coords
sns.scatterplot(data=clusteringDf, x="LONGITUD", y="LATITUD")
```

The cluster model gets initialized 3 clusters because the dataset labels each entry based on the semaphore column

```python
# creating instance of KMeans model
cluster = KMeans(n_clusters=3)

# training model
cluster.fit(clusteringDf)

# predicting model
y_pred = cluster.predict(clusteringDf)
```

```python
clusteringDf['K_MEANS_PRED'] = y_pred
```

```python
clusteringDf['SEMAFORO'] = cleanedDf['SEMAFORO']
```

```python
clusteringDf.head()
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,7)) 
sns.scatterplot(
    data=clusteringDf, 
    x="LONGITUD", 
    y="LATITUD", 
    hue="K_MEANS_PRED",
    ax = ax1
).set(title='KMeans clustering')
sns.scatterplot(
    data = clusteringDf,
    x = "LONGITUD",
    y = "LATITUD",
    hue = "SEMAFORO",
    ax = ax2,
    palette=dict(Verde="green", Amarillo="yellow", Rojo="red")
).set(title='Semaphore label')
```

Based on the illustrations shown above, it can be concluded that there is no relationship between entry coordinates and water quality
